```python
# Scrape golf scores for live tournaments from https://www.livesport.com/en/golf/
import time, sys, os
import pandas as pd
from selenium import webdriver
from selenium.webdriver import Chrome
from bs4 import BeautifulSoup

er = '1'            #1/2/3
tour = 'PGA'        #PGA / EUR
live = 0            #0/1
date = '2023_1_5'
link = 'https://www.livesport.com/en/golf/pga-tour/3m-open/'

```

```python
# 1) Scrape, sort & output scores
driver = Chrome()
driver.get(link)
time.sleep(5)
pageTournLdrbd = driver.page_source
cookies = driver.find_element_by_css_selector("button[id='onetrust-accept-btn-handler']")
cookies.click()
soup = BeautifulSoup(pageTournLdrbd, 'html.parser') #Scrape & Parse Leaderboard Summary
table = soup.find_all('div', {'class':'sportName golf'})
if len(table) == 1:
        rows_id = table[0].find_all('div', {'class':'event__match event__match--noDuel'})
        rows_idf = table[0].find_all('div', {'class':'event__match event__match--last event__match--noDuel'})
print('\nlen(soup) = '+str(len(soup)))
print('len(table) = '+str(len(table)))
print('len(rows_id) = '+str(len(rows_id)))
print('len(rows_idf) = '+str(len(rows_idf)))
```
